# Complex Even vs. Odd Functions

Let's look at more complex polynomials! Instead of just single terms ($y=x^2$, $y=x^3$), let's see what happens when we mix terms.

### 💡 The Golden Rule of Polynomial Symmetry:
- If a polynomial contains **ONLY EVEN degree terms** (remember: a constant like $+9$ is technically $9x^0$, which is even), the entire function is an **Even Function** and is perfectly symmetric about the y-axis.
- If a polynomial contains **ONLY ODD degree terms**, the entire function is an **Odd Function** and has point-symmetry about the origin $(0,0)$.
- *(If a polynomial has a mix of both, it is **Neither**! It loses its perfect symmetry.)*

## 1. Even Functions (y-axis Symmetry)
Notice how the turning points (local maximums and minimums) are perfectly mirrored across the y-axis!

```python
import numpy as np
import plotly.graph_objects as go

x = np.linspace(-3.5, 3.5, 500)

# ==========================================
# Complex Even Polynomials
# ==========================================
def even1(x): return x**4 - 10*x**2 + 9    # Roots at -3, -1, 1, 3
def even2(x): return -x**4 + 4*x**2 + 5    # M-shaped 

fig_even = go.Figure()
fig_even.add_trace(
    go.Scatter(x=x, y=even1(x), name='<i>y</i> = <i>x</i><sup>4</sup> - 10<i>x</i><sup>2</sup> + 9', line=dict(color='#1f77b4', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>')
)
fig_even.add_trace(
    go.Scatter(x=x, y=even2(x), name='<i>y</i> = -<i>x</i><sup>4</sup> + 4<i>x</i><sup>2</sup> + 5', line=dict(color='#9467bd', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>')
)

fig_even.update_layout(
    title_text="Even Functions (All terms are even degree)",
    title_x=0.5,
    title_font_size=20,
    plot_bgcolor='white',
    hovermode="x unified",
    width=900,
    height=500,
    legend=dict(yanchor="top", y=-0.15, xanchor="center", x=0.5, orientation="h")
)

fig_even.update_yaxes(range=[-20, 20])
fig_even.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', title_text="x")
fig_even.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', title_text="y")

fig_even.show()
```

## 2. Odd Functions (Origin Symmetry)
Now look at these functions. If you take a point in the top right (Quadrant I), you will find its exact opposite in the bottom left (Quadrant III). For every $(x, y)$, there is a point $(-x, -y)$!

```python
import numpy as np
import plotly.graph_objects as go

x = np.linspace(-3.5, 3.5, 500)

# ==========================================
# Complex Odd Polynomials
# ==========================================
def odd1(x): return x**3 - 9*x
def odd2(x): return -0.5*x**5 + 4*x**3 - 7*x 

fig_odd = go.Figure()
fig_odd.add_trace(
    go.Scatter(x=x, y=odd1(x), name='<i>y</i> = <i>x</i><sup>3</sup> - 9<i>x</i>', line=dict(color='#ff7f0e', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>')
)
fig_odd.add_trace(
    go.Scatter(x=x, y=odd2(x), name='<i>y</i> = -0.5<i>x</i><sup>5</sup> + 4<i>x</i><sup>3</sup> - 7<i>x</i>', line=dict(color='#d62728', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>')
)

fig_odd.update_layout(
    title_text="Odd Functions (All terms are odd degree)",
    title_x=0.5,
    title_font_size=20,
    plot_bgcolor='white',
    hovermode="x unified",
    width=900,
    height=500,
    legend=dict(yanchor="top", y=-0.15, xanchor="center", x=0.5, orientation="h")
)

fig_odd.update_yaxes(range=[-20, 20])
fig_odd.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', title_text="x")
fig_odd.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', title_text="y")

fig_odd.show()
```

## 3. Neither (Mixed Degrees - No Symmetry)

If you mix even a single Odd term into an Even function, or an Even term into an Odd function, **all symmetry is completely destroyed**!

We call these functions **Neither**. Notice how their left side doesn't match their right side in any way (no mirror, no 180° rotation match), although their *End Behavior* is still completely controlled by the Leading Term!

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

x = np.linspace(-4, 4, 500)

# ==========================================
# Mixed Degree Polynomials (Neither)
# ==========================================
def mixed1(x): return x**3 - 4*x**2 + 2    # Odd leading (x^3), but has Even term (x^2, constant)
def mixed2(x): return -0.5*x**4 + 2*x**3 + 4*x**2 # Even leading (x^4), but has Odd term (x^3)

fig_mixed = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        "<b>Mixed Degree (Odd Leading)</b>", 
        "<b>Mixed Degree (Even Leading)</b>"
    ),
    horizontal_spacing=0.1
)

fig_mixed.add_trace(
    go.Scatter(x=x, y=mixed1(x), name='<i>y</i> = <i>x</i><sup>3</sup> - 4<i>x</i><sup>2</sup> + 2', line=dict(color='#2ca02c', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>'),
    row=1, col=1
)

fig_mixed.add_trace(
    go.Scatter(x=x, y=mixed2(x), name='<i>y</i> = -0.5<i>x</i><sup>4</sup> + 2<i>x</i><sup>3</sup> + 4<i>x</i><sup>2</sup>', line=dict(color='#8c564b', width=3),
               hovertemplate='x: %{x:.2f}<br>y: %{y:.2f}<extra></extra>'),
    row=1, col=2
)

fig_mixed.update_layout(
    title_text="Functions with Mixed Parity (Neither)",
    title_x=0.5,
    title_font_size=20,
    plot_bgcolor='white',
    hovermode="x unified",
    width=1000,
    height=500,
    legend=dict(yanchor="top", y=-0.15, xanchor="center", x=0.5, orientation="h")
)

for i in [1, 2]:
    fig_mixed.update_yaxes(range=[-30, 20], row=1, col=i)
    fig_mixed.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', row=1, col=i, title_text="x")
    fig_mixed.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', row=1, col=i, title_text="y")

fig_mixed.show()
```
