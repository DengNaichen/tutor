# Why does the Leading Term determine End Behavior?

In this notebook, we visualize why we only need to look at the leading term of a polynomial to determine its end behavior. 

We use the example $f(x) = x^3 - 50x^2 + 200$ and compare it with its leading term $y = x^3$.

**This is an interactive plot using Plotly!**
- **Hover** your mouse over the lines to see the exact (x, y) coordinates.
- **Click and drag** to zoom into any specific region.
- **Double-click** to reset the zoom.

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Define the functions
def f(x):
    return x**3 - 50*x**2 + 200

def g(x):
    return x**3

# Pre-calculate data for smooth rendering
x_local = np.linspace(-15, 60, 500)
x_end = np.linspace(-100000, 100000, 500)

# Create a subplot figure with 1 row and 2 columns
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=("Local Behavior (Zoomed In)", "End Behavior (Zoomed Out)"),
    horizontal_spacing=0.1
)

# ==========================================
# Panel 1: Local Behavior
# ==========================================
fig.add_trace(
    go.Scatter(
        x=x_local, y=f(x_local), 
        name='f(x) = x^3 - 50x^2 + 200',
        line=dict(color='#1f77b4', width=3),
        hovertemplate='x: %{x:.1f}<br>y: %{y:.1f}<extra></extra>' # <extra></extra> hides trace name in hover
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=x_local, y=g(x_local), 
        name='Leading term: y = x^3',
        line=dict(color='#ff7f0e', width=3, dash='dash'),
        hovertemplate='x: %{x:.1f}<br>y: %{y:.1f}<extra></extra>'
    ),
    row=1, col=1
)

# ==========================================
# Panel 2: End Behavior
# ==========================================
fig.add_trace(
    go.Scatter(
        x=x_end, y=f(x_end), 
        name='f(x) = x^3 - 50x^2 + 200',
        showlegend=False, # Don't duplicate legend items
        line=dict(color='#1f77b4', width=6), # Thicker to show overlap better
        opacity=0.6,
        hovertemplate='x: %{x:.1f}<br>y: %{y:.1e}<extra></extra>'
    ),
    row=1, col=2
)

fig.add_trace(
    go.Scatter(
        x=x_end, y=g(x_end), 
        name='Leading term: y = x^3',
        showlegend=False,
        line=dict(color='#ff7f0e', width=3, dash='dash'),
        hovertemplate='x: %{x:.1f}<br>y: %{y:.1e}<extra></extra>'
    ),
    row=1, col=2
)

# ==========================================
# Layout Updates
# ==========================================
fig.update_layout(
    title_text="Why the Leading Term Determines End Behavior",
    title_x=0.5, # Center title
    title_font_size=20,
    plot_bgcolor='white',
    hovermode="x unified", # Shows hover info for all lines at the same x position!
    width=1000,
    height=600,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.2,
        xanchor="center",
        x=0.5
    )
)

# Add axes lines and grid to both subplots
for i in [1, 2]:
    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', row=1, col=i, title_text="x")
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGray', zeroline=True, zerolinewidth=2, zerolinecolor='Black', row=1, col=i, title_text="y")

fig.show()
```
